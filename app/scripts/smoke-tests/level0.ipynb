```python
import sys
import numpy as np
import xarray as xr

sys.path.insert(0, "/home/stefan/CRS/CRS.ZarrPipelines/")
from app.domain.scoring import score_zarr_multi, score_zarr_minmax
from app.utils.scoring_config import ScoringConfig
```

```python
cfg = ScoringConfig()

# Tiny synthetic DataArray — 10×10 grid, values 0–50
data = xr.DataArray(np.linspace(0, 50, 100).reshape(10, 10), dims=["lat", "lon"])

thresholds_per_scale = {
    "5":  cfg.get_thresholds("HS", "5"),   # first metric auto-selected
    "10": cfg.get_thresholds("HS", "10"),
}
```

```python
# Threshold-based scoring (5-point and 10-point)
scored = score_zarr_multi(data, scales=["5", "10"], thresholds_per_scale=thresholds_per_scale)
print(scored)
print("5-point:\n",  scored.sel(scoring="5").score.values)
print("10-point:\n", scored.sel(scoring="10").score.values)

# Min-max 0–100 scoring
minmax = score_zarr_minmax(data)
print("\n0-100 min-max:\n", minmax.score.values)
```
